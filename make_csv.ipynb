```python
import shutil, os, random, zipfile, math
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
# 폴더별 경로 및 파일 갯수 확인
dir = '/home/ubin/Project/jordan'
class_path = []
cname = []
fname = []
for dirpath, class_name, filename in os.walk(dir):
    class_path.append(dirpath)
class_path2 = []
fname = []
for dir in class_path[1:]:
    for dirpath, class_name, filename in os.walk(dir):
        class_path2.append(dirpath)
        fname.append(filename)
for i in range(len(fname)):
    print(class_path2[i], '\n file number:', len(fname[i]))
```

```python
# 클래스별 train, test 데이터 갯수 확인
for i in range(len(fname)):
    print(
    '클래스 :', class_path2[i].split('/')[-1], 
    '\n 클래스별 전체 data 갯수 :', len(fname[i]),
    '\n 클래스별 train data 갯수 :', math.floor(len(fname[i]) * 0.8),
    '\n 클래스별 test data 갯수 :', len(fname[i])-math.floor(len(fname[i]) * 0.8))
```

```python
# 전체 파일 경로 생성
path = []
for i in range(len(class_path2)):
    for j in range(len(fname[i])):
        path.append(os.path.join(class_path2[i], fname[i][j]))
```

```python
# 경로에 맞는 클래스명 생성
class_name = []
for i in range(len(path)):
    class_name.append(path[i].split('/')[-2])
```

```python
# 전체 파일 csv 만들기
df_path = pd.DataFrame(data = path, columns=['path'])
df_class = pd.DataFrame(data = class_name, columns=['class_name'])
df_shoes = df_path.join(df_class)
df_shoes
```

```python
# train, test 클래스별 8:2로 data 나눔
shoes_train, shoes_test = train_test_split(df_shoes, test_size=0.2, random_state=123, stratify=class_name)
print(shoes_train,'\n', shoes_test)
```

```python
# 각 클래스별 8:2로 나누어 졌는지 train, test 데이터프레임 shape 확인
unique_class = np.unique(class_name)
for i in unique_class:
    print(i, '\ntrain :', shoes_train[shoes_train.class_name==i].shape,
    '\ntest :', shoes_test[shoes_test.class_name==i].shape)
```

```python
# 분류가 제대로 되었는지 다시 한번 체크
for i in range(len(class_path2)):
    print(class_path2[i].split('/')[-1],
    '\ntrain :', np.round(len(fname[i])*0.8), '\ntest :', np.round(len(fname[i])*0.2))
```
